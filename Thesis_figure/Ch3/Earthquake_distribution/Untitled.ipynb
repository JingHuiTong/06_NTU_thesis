```python
import os, glob
import pandas as pd
```

```python
path = '/Volumes/home/Research/STEP/02_Station_result_csv'
```

```python
newdf = {'Time':[],
        'Lon':[],
        'Lat':[],
        'Dep':[],
        'Mag':[] }
for csvpath in glob.glob(f'{path}/*classify_220913/*v2.csv'):
    df = pd.read_csv(csvpath)
    dff = df[df['Pick']==True]
    for i in range(len(dff)):
        time  = dff['Event'].values[i]
        evlat = dff['Ev_lat'].values[i]
        evlon = dff['Ev_lon'].values[i]
        depth = dff['Depth'].values[i]
        mag   = dff['Mag'].values[i]
        if time  not in newdf['Time'] :
            if evlon  not in newdf['Lon']:
                newdf['Time'].append(time)
                newdf['Lat'].append(evlat)
                newdf['Lon'].append(evlon)
                newdf['Dep'].append(depth)
                newdf['Mag'].append(mag)
     

dfff = pd.DataFrame(newdf) 
dfff.to_csv('/Volumes/home/06_NTU_thesis/Thesis_figure/Ch3/Earthquake_distribution/Event_used_2010-2020_221013.csv',index=False)
```
