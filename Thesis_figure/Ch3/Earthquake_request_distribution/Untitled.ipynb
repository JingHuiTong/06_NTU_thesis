```python
import obspy 
from obspy.clients.fdsn import Client
from obspy import UTCDateTime
from obspy.geodetics.base import gps2dist_azimuth , kilometer2degrees
client = Client("IRIS")
import glob, os
import pandas as pd
```

```python
newdf = {'Time':[],
        'Lon':[],
        'Lat':[],
        'Deg':[],
        'Depth':[],
        'Mag':[]}
for csv in glob.glob('*csv'):
    df = pd.read_csv(csv)
    for i in range(len(df)):
        time  = df['Time'].values[i]
        evlon = df['Lon'].values[i]
        evlat = df['Lat'].values[i]
        mag   = df['Mag'].values[i]
        depth = df['Depth'].values[i]/1000
        dist, az, baz = gps2dist_azimuth(41.115,43.8036, evlat, evlon)
        deg = kilometer2degrees(dist/1000)
        
        if 85<=deg<=150 and time < UTCDateTime('20201231T23:59:59'):
            newdf['Time'].append(time)
            newdf['Lat'].append(evlat)
            newdf['Lon'].append(evlon)
            newdf['Mag'].append(mag)
            newdf['Depth'].append(depth)
            newdf['Deg'].append(deg)            

            
newdf = pd.DataFrame(newdf)
newdf = newdf.sort_values(by=['Time'],ignore_index=True)
newdf.to_csv('Earthquake_clear_85-150_2010-2020.csv',index=False)
```
