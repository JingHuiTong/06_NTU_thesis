```python
 import pickle
import obspy
import matplotlib.pyplot as plt
```

```python
Ani = 'version3/baz_300/LQT_data.pkl'
Iso = 'version3/baz_80/LQT_data.pkl'
```

```python
Ani = pickle.load(open(Ani, "rb"))
Iso = pickle.load(open(Iso, "rb"))
```

```python
Ani_R = Ani.select(component='Q')[0]
Ani_T = Ani.select(component='T')[0]
Iso_R = Iso.select(component='Q')[0]
Iso_T = Iso.select(component='T')[0]
```

```python
time = Ani_R.times()
maxP = max(Iso_R.data)
```

```python
xmin = 40
plt.figure(figsize=(5,4))
plt.subplot(212)
plt.plot(time,Ani_T.data/maxP,'k',lw=2)
plt.plot(time,Ani_R.data/maxP,'purple',lw=2)
plt.text(xmin+3,1.2,'R',ha='left',va='top', color='purple',fontsize=22,fontweight='bold')
plt.text(xmin+3,-1.2,'T',ha='left',va='bottom', color='k',fontsize=22,fontweight='bold')
plt.xlim(xmin,xmin+35)
plt.ylim(-1.4,1.4)
plt.yticks([])
# plt.text(xmin+43,-1.2,'Anisotropy',ha='right',color='b',fontsize=15,fontweight='bold',style='italic')
plt.subplot(211)
plt.plot(time,Iso_T.data/maxP,'k',lw=2)
plt.plot(time,Iso_R.data/maxP,'purple',lw=2)
plt.text(xmin+3,1.2,'R',ha='left',va='top' ,color='purple',fontsize=22,fontweight='bold')
plt.text(xmin+3,-1.2,'T',ha='left',va='bottom', color='k',fontsize=22,fontweight='bold')
# plt.text(xmin+43,-1.2,'Isotropy',ha='right',color='b',fontsize=15,fontweight='bold',style='italic')
plt.xlim(xmin,xmin+35)
plt.yticks([])
plt.ylim(-1.5,1.4)
plt.savefig('Anisotropy_example_waveform_v3.png',dpi=200)
plt.savefig('Anisotropy_example_waveform_v3.pdf')
```
