```python
 import pickle
import obspy
from obspy import UTCDateTime
import matplotlib.pyplot as plt
```

```python
Ani = 'baz_270/LQT_data.pkl'
Ani_PM = 'baz_270/ZNE_data.pkl'
Iso = 'baz_80/LQT_data.pkl'
Iso_PM = 'baz_80/ZNE_data.pkl'
```

```python
Ani = pickle.load(open(Ani, "rb"))
Iso = pickle.load(open(Iso, "rb"))
Ani_PM = pickle.load(open(Ani_PM, "rb"))
Iso_PM = pickle.load(open(Iso_PM, "rb"))
```

```python
Ani_R = Ani.select(component='Q')[0]
Ani_T = Ani.select(component='T')[0]
Iso_R = Iso.select(component='Q')[0]
Iso_T = Iso.select(component='T')[0]

```

```python
time = Ani_R.times()
maxP = max(Ani_R.data)
trstart = UTCDateTime('2022-03-04T17:18:55.000000Z')
starttime = UTCDateTime('2022-03-04T17:19:43.000000Z')
endtime = UTCDateTime('2022-03-04T17:20:08.000000Z')
```

```python
xmin = 30
plt.figure(figsize=(5,4))
plt.subplot(211)
# plt.fill_betweenx(x1=starttime-trstart, x2=endtime-trstart, y=[-1.5,1.5],color='grey',alpha=0.2)
plt.plot(time,Ani_T.data/maxP,'k')
plt.plot(time,Ani_R.data/maxP,'purple')
plt.text(xmin+3,1,'R',ha='left',va='top', color='purple',fontsize=12)
plt.text(xmin+3,-1.2,'T',ha='left',va='top', color='k',fontsize=12)
plt.xlim(xmin,xmin+55)
plt.ylim(-1.6,1.2)
plt.yticks([])
plt.text(xmin+53,-1.2,'Anisotropy',ha='right',color='b',fontsize=15,fontweight='bold',style='italic')
plt.subplot(212)
plt.plot(time,Iso_T.data/maxP,'k')
plt.plot(time,Iso_R.data/maxP,'purple')
plt.text(xmin+3,1,'R',ha='left',va='top' ,color='purple',fontsize=12)
plt.text(xmin+3,-1.2,'T',ha='left',va='top', color='k',fontsize=12)
plt.text(xmin+53,-1.2,'Isotropy',ha='right',color='b',fontsize=15,fontweight='bold',style='italic')
plt.xlim(xmin,xmin+55)
plt.ylim(-1.6,1.2)
plt.yticks([])
# plt.savefig('Anisotropy_example_waveform.png',dpi=200)
# plt.savefig('Anisotropy_example_waveform.pdf')
```
